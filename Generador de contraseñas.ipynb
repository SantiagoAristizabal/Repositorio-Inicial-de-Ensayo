```python
import numpy as np
from random import shuffle
import pandas as pd
from pandas import DataFrame as DF
```

```python
def generar_contraseña():
    minúscula = []
    mayúscula = []
    número = []
    símbolo = []
    casi_cont = []
    comando_semifinal = []
    minus = ["qwertyuiopasdfghjklzxcvbnm"]
    mayus = ["QWERTYUIOPASDFGHJKLZXCVBNM"]
    num = ["6084315972"]
    simb = ["|!)#${%&/.,(=*<}-:_+@];.?[>"]
    ele = ["minus","mayus","num","simb"]
    elementos = ["minúscula","mayúscula","número","símbolo"]
    E = list(zip(ele,elementos))
    for i,ii in E: exec("for i in range(len(" + i + "[0])):" + ii + ".append(" + i + "[0][i])")
    for i,ii in E: exec("len" + i + "=" + "len(" + ii + ")" )
    for i,ii in E: exec("enu" + i + "=" + "list(enumerate(" + ii + "))" )
    for i in ele: exec("cant_elemento_" + i + "=" + "np.random.randint(low = 4, high = 6+1)" )
    for i in ele: exec("ind" + i + " = " + "np.random.randint(len" + i + ", size" + " = " + "cant_elemento_" + i +")")
    for i in ele: exec("casi_contraseña_" + i + "=[]")
    for i in ele: exec("for i in ind" + i +": casi_contraseña_" + i + ".append(enu" + i + "[i][1])")
    for i in ele: exec("casi_cont.append(casi_contraseña_" +i+")")
    for i in range(len(casi_cont)):
        if i == len(casi_cont)-1:
            comando_semifinal.append("''.join(casi_cont[" + str(i) + "]) "  )
        else:
            comando_semifinal.append("''.join(casi_cont[" + str(i) + "]) +"  )
    comando_final = "".join(comando_semifinal)
    final = list((eval(comando_final)))
    #Revolver 15 veces
    exec("shuffle(final),"*6 + "shuffle(final)")
    contraseña = ''.join(final)
    return(contraseña)

def generar_contraseñas(número_contraseñas = 10):
    ncm1 = número_contraseñas + 1
    contraseñas = []
    for i in range(número_contraseñas): contraseñas.append(generar_contraseña())
    datos = DF(contraseñas,columns=["Contraseñas generadas"],index=range(1,ncm1))
    return(datos)
```
